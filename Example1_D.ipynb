```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Restaurant": ["A", "B", "C", "D", "E"],
    "Rating": [4.1, 3.8, 4.3, 3.5, 4.0],
    "Votes": [120, 90, 200, 50, 160],
    "Online_Order": ["Yes", "No", "Yes", "No", "Yes"],
    "Type": ["Dining", "Cafe", "Delivery", "Dining", "Cafe"],
    "Cost_for_two": [300, 200, 400, 250, 300]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.head()
```

sns.countplot(x=df["Type"])
plt.title("Restaurant Types")
plt.show()

```python
sns.countplot(x=df["Online_Order"])
plt.title("Online Order Availability")
plt.show()
```

```python
plt.hist(df["Rating"], bins=5)
plt.title("Rating Distribution")
plt.show()
```

```python
sns.barplot(x=df["Restaurant"], y=df["Cost_for_two"])
plt.title("Cost for Two People")
plt.show()
```
