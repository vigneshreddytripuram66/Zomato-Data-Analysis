```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("IPL_Squad_2023_Auction_Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True, errors='ignore')
```

```python
df.isnull().sum()
```

```python
costliest = df.sort_values(by="COST IN ₹ (CR.)", ascending=False)

costliest.head(10)
```

```python
team_spending = df.groupby("Team")["COST IN ₹ (CR.)"].sum()

team_spending.plot(kind="bar")

plt.title("Total Money Spent by Each Team")
plt.xlabel("Team")
plt.ylabel("Cost in Crores")
plt.show()
```

```python
sns.countplot(x="TYPE", data=df)

plt.title("Distribution of Player Types")
plt.show()
```

```python
unsold_players = df[df["COST IN ₹ (CR.)"] == 0]

print("Total Unsold Players:", len(unsold_players))
```

```python
sns.boxplot(x="TYPE", y="COST IN ₹ (CR.)", data=df)

plt.title("Cost Distribution by Player Type")
plt.show()
```

```python
team_spending.sort_values(ascending=False).head(1)
```

```python
type_counts = df["TYPE"].value_counts()

type_counts.plot(kind="pie", autopct="%1.1f%%")

plt.title("Player Type Percentage")
plt.show()
```
